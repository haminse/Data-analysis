```python
#기준 연월 컬럼 생성
df['기준연월'] = '2019-01'
df
```

```python
df['국적'].unique()
```

```python
# 제외하고자 하는 목록 만들기
del_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계']

df_country = df[df['국적'].isin(del_list) == False]
df_country
df_country['국적'].unique()
```

```python
df_country = df_country.reset_index(drop = True)
df_country
```

```python
# continents 축 만들기용 series
continents = ['아시아']*25 + ['아메리카']*5 + ['유럽']*23 + ['오세아니아']*3 + ['아프리카']*2 + ['기타대륙'] + ['교포']
print(continents)
```

```python
df_country['대륙'] = continents
df_country
```

```python
#관광객 비율 컬럼
df_country['관광객 비율(%)'] =round( (df_country['관광']/df_country['계']) *100, 1)
df_country
```

```python
df_country.sort_values(by = '관광객 비율(%)', ascending = 'True')
```

```python
df_country.pivot_table(index = '대륙', values = ['관광객 비율(%)'], aggfunc = 'mean')
#이렇게 하면 오류가 발생 - 평균(관광객 비율)의 평균을 계산한 셈이 된다
```
